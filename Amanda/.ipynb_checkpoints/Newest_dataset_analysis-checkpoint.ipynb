```python

cols = ['d18O', 'd2H', 'd13C', 'd34S']
cmaps = ['Blues', 'Greens', 'Reds', 'Purples']  

fig, axes = plt.subplots(2, 2, figsize=(10, 6))

for ax, col, cmap in zip(axes.flatten(), cols, cmaps):
    usa.plot(ax=ax, color='lightgray', edgecolor='black')
    
    tree.plot(
        column=col,
        legend=True,
        s=30,
        ax=ax,
        cmap=cmap,
        edgecolor='black',
        linewidth=0.3,  
        legend_kwds={'label': f"{col}"}
    )
    
    ax.set_title(f"{col} for all trees sampled")
    ax.set_xlabel('Latitude')
    ax.set_ylabel('Longitude')

plt.tight_layout()
plt.show()
```

```python
# Calculate deterium excess 
tree['d_excess'] = tree.d2H - 8*(tree.d18O)

#tree['d_excess']
```

```python
#create a plot that looks at tree diameter and respective isotope measurements

cols = ['d18O', 'd2H', 'd13C', 'd34S']

fig, axes = plt.subplots(2, 2, figsize=(12, 10), sharex=True)

for ax, col in zip(axes.flatten(), cols):
    df_sub = tree[['Diameter of tree at sample location (inches)', col]].dropna()
    X = df_sub[['Diameter of tree at sample location (inches)']]
    y = df_sub[col]
    
    # Fit regression
    model = LinearRegression().fit(X, y)
    y_pred = model.predict(X)
    r2 = r2_score(y, y_pred)
    
    # Scatter plot
    sns.scatterplot(x='Diameter of tree at sample location (inches)', y=col, data=df_sub, ax=ax, alpha=0.7)
    
    # Regression line
    ax.plot(X, y_pred, color='red')
    
    # Annotate R²
    ax.text(0.05, 0.95, f"$R^2 = {r2:.2f}$", 
            transform=ax.transAxes, 
            fontsize=12, 
            verticalalignment='top',
            bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.6))
    
    ax.set_title(f"{col} vs Diameter of tree at sample location (inches)")

plt.tight_layout()
plt.show()
```

## Create linear models to predict tree isotopes given single parameters. Then, record and plot their r2 performance. 

# Climate Normals

```python
#isotopes 
isotopes = ['d18O', 'd2H', 'd13C', 'd34S', 'd_excess']

#variables of interest 
clim_normal_predictors = [

 'annual_precipitation_PRISIM',
 'annual_mean_temperature_PRISIM', 
 'annual_maximum_temperature_PRISIM',
 'annual_minimum_temperature_PRISIM',
 'annual_tmin_nClimGrid', 
    'annual_tavg_nClimGrid',
     'annual_tmax_nClimGrid',
     'annual_precipitation_nClimGrid', 
    
]
      
#compute r2 
results = []

for iso in isotopes:
    for var in clim_normal_predictors:
        df_pair = tree[[var, iso]].dropna()
        ds_length = df_pair.shape[0]

        X = df_pair[[var]].values
        y = df_pair[iso].values

        model = LinearRegression().fit(X, y)
        r2 = r2_score(y, model.predict(X))

        results.append({
            'Isotope': iso,
            'Variable': var,
            'R2': r2,
            'ds_length': ds_length # to ensure we are comparing apples to apples (and the dataset lengths are similar)
        })

#make table 
r2_table_clim_normal = pd.DataFrame(results) \
    .sort_values(['Isotope', 'R2'], ascending=[True, False]) \
    .reset_index(drop=True)

r2_table_clim_normal.head()

r2_table_clim_normal_sorted = r2_table_clim_normal.sort_values(by='R2', ascending=False).reset_index(drop=True)

r2_table_clim_normal_sorted.head(3)
```

```python
#look at top 10 for climate normals 

top_r2_per_isotope_clim_norm = (
    r2_table_clim_normal_sorted
    .sort_values(['Isotope', 'R2'], ascending=[True, False])
   .groupby('Isotope')
    .head(10)
   .reset_index(drop=True)
)

top_r2_per_isotope_clim_norm.head()

```

```python
# Map isotopes w good labels 
iso_labels = {
    'd18O': 'δ¹⁸O',
    'd2H': 'δ²H',
    'd13C': 'δ¹³C',
    'd34S': 'δ³⁴S', 
    'd_excess': 'd_excess', 
}

# Define a consistent color palette for all variables
unique_vars = top_r2_per_isotope_clim_norm['Variable'].unique()
palette = sns.color_palette('tab20', n_colors=len(unique_vars))
var_color_map = dict(zip(unique_vars, palette))

# Set up plot
fig, axes = plt.subplots(1, 5, figsize=(25, 10), sharey=False)
axes = axes.flatten()

# Get unique isotopes
isotopes = top_r2_per_isotope_clim_norm['Isotope'].unique()

# Plot each isotope
for i, iso in enumerate(isotopes):
    data = top_r2_per_isotope_clim_norm[top_r2_per_isotope_clim_norm['Isotope'] == iso]
    sns.barplot(
        data=data,
        x='Variable',
        y='R2',
        palette=[var_color_map[v] for v in data['Variable']],
        ax=axes[i]
    )
    axes[i].set_ylim(0,0.8)
    axes[i].set_title(f'{iso_labels[iso]}', fontsize=20)
    axes[i].set_xlabel('Predictor Variable', fontsize=20)
    axes[i].set_ylabel('$R^2$', fontsize=20)
    axes[i].tick_params(axis='x', rotation=90, labelsize=20)
    axes[i].tick_params(axis='y', labelsize=16)

# Add overall title
plt.suptitle("Linear Regression Model Performance ($R^2$) by Isotope \nClimate Normal parameters", fontsize=28, y=1.03)

# Final layout
plt.tight_layout()
plt.show()
```
